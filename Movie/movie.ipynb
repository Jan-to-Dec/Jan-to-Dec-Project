# 데이터 전처리 data preprocessing

```python
movies_df = movies[['id', 'budget', 'genres', 'title', 'release_date',
                    'revenue', 'vote_average', 'vote_count']]
credits_df = credits[['movie_id', 'cast', 'crew']]
```

```python
data = pd.merge(movies_df, credits_df, left_on  = 'id', right_on = 'movie_id').drop('movie_id', axis=1)
data.head()
```

### Create ROI column 

```python
data['roi'] = data['revenue'] / data['budget']
data.head()
```

### Create Director column

```python
data['crew']
```

```python
print(ast.literal_eval(data['crew'][0]))
```

```python
data['crew'] = data['crew'].apply(ast.literal_eval)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director' :
            return i['name']
```

```python
data['director'] = data['crew'].apply(get_director)
```

```python
data.head()
```

### Create cast_name column

```python
data['cast']
```

```python
data['cast_name'] = list(map(lambda x: [i['name'] for i in ast.literal_eval(x)], data['cast']))
data.head()
```

### Create genres column

```python
data['genres']
```

```python
data['genres'] = data['genres'].apply(ast.literal_eval)
```

```python
def get_genres(x):
    if len(x) > 0:
        return x[0]['name']
data['main_genre'] = data['genres'].apply(get_genres)
data.head()    
```

### Data Type Conversion

```python
data['release_date'] = pd.to_datetime(data['release_date'], format='%Y-%m-%d')
data['id'] = data['id'].astype(str)
```

### Create Year, Month columns

```python
data['year'] = data['release_date'].dt.year
data['month'] = data['release_date'].dt.month
```

### 결측치 제거

```python
data.info()
```

```python
data.dropna(inplace=True)
```

### 2. Top 10 highest-grossing movies

```python
top = data.groupby('title')['revenue'].sum().reset_index().sort_values('revenue', ascending=False).head(10)
plt.figure(figsize=(10, 6)) 
plt.bar(top['title'], top['revenue'])
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

예산 TOP 10, 투표수 ,TOP 10 영화

```python
import plotly.express as px  

title_dic = {'budget': '예산', 'vote_count': '투표수'}
for y in ['budget', 'vote_count']:
    top = data.groupby('title')[[y]].sum().reset_index().sort_values(y, ascending=False).head(10)
    fig = px.bar(data_frame=top, x='title', y=y, title=title_dic[y] + " Top 10")  
    fig.show()
```

예산에 따라, 투표수에 따라 흥생 수익도가 얼마나 차이가 나는지 볼 수 있었다. 각각은 정비례하지 않아서 상관관계를 알아볼 필요가 있다.

### 4. Distribution of Box Office Revenue by Genre
- 흥행 수익 좋은 장르
- 시간의 흐름에 따라 장르가 바뀌는지에 대한 여부
- 월별로 흥행하는 장르


```python
# 장르별 흥행 수익 분포
fig = px.box(data_frame = data, y = 'main_genre', x = 'revenue', hover_name = 'title')
fig.show()
```

액션과 어드벤처 장르가 가장 높은 수익을 기록하며, 특히 일부 작품들은 압도적인 흥행을 보였습니다. 외국 TV, 음악, 다큐멘터리 등 일부 장르는 전체적으로 낮은 수익을 기록했습니다.
중앙값을 비교하면 다른 장르에 비해 높은 편은 아닙니다. 

```python
data.groupby('main_genre')[['revenue']].mean().reset_index()
```

```python
avg_genre_rev = data.groupby('main_genre')[['revenue']].mean().reset_index()
fig = px.bar(data_frame = avg_genre_rev, x = 'main_genre', y = 'revenue', title = ' 징르별 흥행 수익 평균')
fig.show()
```

장르별 영화 수익 평균 순위는 Animation > Adventure > Family > SF > Fantasy > Action이다. 

```python
# 연도별로 나타난 각각의 장르 수익
revenue_by_year_genre = data.query('year >= 1990').groupby(['year', 'main_genre'])[['revenue']].sum().reset_index()
```

```python
fig = px.bar(
    data_frame=revenue_by_year_genre,
    x="year",
    y="revenue",
    color='main_genre',
    color_discrete_sequence=px.colors.qualitative.Plotly,
    opacity=0.7
)

fig.show()
```

```python
revenue_by_month_genre = data.query('year >= 1990').groupby(['month', 'main_genre'])[['revenue']].sum().reset_index()
```

```python
fig = px.bar(
    data_frame=revenue_by_month_genre,
    x="month",
    y="revenue",
    color='main_genre',
    color_discrete_sequence=px.colors.qualitative.Plotly,
    opacity=0.7
)

fig.show()
```

Action과 Adventure는 봄과 여름에 상대적으로 수익이 높았습니다. 
Drama는 가을과 겨울에 수익이 높았고, comedy는 겨울에 수익이 높았습니다.

```python
top300 = data.sort_values('revenue', ascending=False).head(300)

sns.set_style("whitegrid")
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(data=top300, x='roi', y='revenue', hue='main_genre', size='revenue',
                          sizes=(20, 500), palette='Paired', alpha=0.7, legend=None)

plt.title('Revenue vs. ROI (Top 300)', fontsize=16)
plt.xlabel('ROI', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.show() 
```

Action은 예산이 많이드는 경우가 많습니다. 그래서 ROI가 높은 편은 아닙니다. 
Drama, comedy, romance가 ROI가 큰 경우가 많습니다.
